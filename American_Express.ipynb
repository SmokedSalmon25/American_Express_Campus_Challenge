```python
# Create columns for batsman statistics
num_batsmen = 12  # Adjust according to your maximum number of batsmen
for i in range(1, num_batsmen + 1):
    match_data[f"Team_1_Batsman_{i}_strike_rate"] = pd.Series([0] * len(match_data))
    match_data[f"Team_1_Batsman_{i}_custom_score"] = pd.Series([0] * len(match_data))

batsman['Fours'] = batsman['Fours'].fillna(0)
batsman['Sixes'] = batsman['Sixes'].fillna(0)

# Iterate through each row in match_data
for j in range(match_data.shape[0]):
    roster_ids = match_data["team1_roster_ids"][j]
    match_date = match_data["match_dt"][j]
    batsman_id_list = roster_ids.split(':')
    i = 1
    for batsman_id in batsman_id_list:
        batsman_id = float(batsman_id)

        # Filter the batsman dataset for the current batsman and match date less than the current match date
        filtered_batsmen = batsman[(batsman['batsman_id'] == batsman_id) & (batsman['match_dt'] < match_date)]

        if not filtered_batsmen.empty:
            # Sort filtered DataFrame by 'match_dt' descending
            filtered_batsmen = filtered_batsmen.sort_values(by='match_dt', ascending=False)

            # Calculate the average strike rate over the last 5 matches (if available)
            if filtered_batsmen.shape[0] >= 5:
                last_5_rows = filtered_batsmen.iloc[:5, :]
            else:
                last_5_rows = filtered_batsmen

            avg_strike_rate = last_5_rows['strike_rate'].mean()
            avg_fours = last_5_rows['Fours'].mean()
            avg_sixes = last_5_rows['Sixes'].mean()

            # Calculate the weighted sum of 4's and 6's over the last 5 matches
            weighted_boundaries = (0.4 * avg_fours) + (0.6 * avg_sixes)

            # Define the custom score as the weighted boundaries
            custom_score = weighted_boundaries

            # Store the calculated scores in match_data
            match_data.loc[j, f"Team_1_Batsman_{i}_strike_rate"] = avg_strike_rate
            match_data.loc[j, f"Team_1_Batsman_{i}_custom_score"] = custom_score
        i += 1

match_data.head()
```

```python
# Calculate the average ground_avg_runs_last15 for each ground_id
ground_avg_dict = df.groupby('ground_id')['ground_avg_runs_last15'].mean()

# Function to fill NaN values based on ground_id
def fill_nan(row):
    if pd.isna(row['ground_avg_runs_last15']):
        avg = ground_avg_dict.get(row['ground_id'])
        return avg if avg is not None else 0
    else:
        return row['ground_avg_runs_last15']

# Apply the function to fill NaN values
df['ground_avg_runs_last15'] = df.apply(fill_nan, axis=1)
```

```python
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import accuracy_score
import optuna
import pandas as pd
import numpy as np

# Assume filtered_df and truths are already defined and are Pandas DataFrame and Series respectively
truths = pd.Series(truths)

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(filtered_df, truths, test_size=0.2, random_state=42)

# Define objective functions for each model using Optuna

def objective_xgb(trial):
    param = {
        'verbosity': 0,
        'objective': 'binary:logistic',
        'eval_metric': 'logloss',
        'booster': 'gbtree',
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 0.5),
        'n_estimators': trial.suggest_int('n_estimators', 100, 500),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 5),
        'subsample': trial.suggest_uniform('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_uniform('colsample_bytree', 0.6, 1.0),
        'gamma': trial.suggest_loguniform('gamma', 1e-8, 1.0),
    }

    skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
    accuracy_list = []

    for train_idx, valid_idx in skf.split(X_train, y_train):
        X_train_fold, X_valid_fold = X_train.iloc[train_idx], X_train.iloc[valid_idx]
        y_train_fold, y_valid_fold = y_train.iloc[train_idx], y_train.iloc[valid_idx]

        model = xgb.XGBClassifier(**param, use_label_encoder=False)
        model.fit(X_train_fold, y_train_fold, eval_set=[(X_valid_fold, y_valid_fold)], early_stopping_rounds=10, verbose=False)

        preds = model.predict(X_valid_fold)
        accuracy = accuracy_score(y_valid_fold, preds)
        accuracy_list.append(accuracy)

    return np.mean(accuracy_list)

def objective_lgb(trial):
    param = {
        'objective': 'binary',
        'metric': 'binary_logloss',
        'boosting_type': 'gbdt',
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 0.5),
        'n_estimators': trial.suggest_int('n_estimators', 100, 500),
        'num_leaves': trial.suggest_int('num_leaves', 20, 100),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'min_child_samples': trial.suggest_int('min_child_samples', 5, 50),
        'subsample': trial.suggest_uniform('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_uniform('colsample_bytree', 0.6, 1.0),
        'lambda_l1': trial.suggest_loguniform('lambda_l1', 1e-8, 1.0),
        'lambda_l2': trial.suggest_loguniform('lambda_l2', 1e-8, 1.0),
        'early_stopping_rounds': 10,  # Specify early stopping directly in params
    }

    skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
    accuracy_list = []

    for train_idx, valid_idx in skf.split(X_train, y_train):
        X_train_fold, X_valid_fold = X_train.iloc[train_idx], X_train.iloc[valid_idx]
        y_train_fold, y_valid_fold = y_train.iloc[train_idx], y_train.iloc[valid_idx]

        # Create LightGBM datasets for training
        train_data = lgb.Dataset(X_train_fold, label=y_train_fold)
        valid_data = lgb.Dataset(X_valid_fold, label=y_valid_fold, reference=train_data)

        # Train the model
        model = lgb.train(param, train_data, valid_sets=[valid_data], num_boost_round=1000)

        # Predict and calculate accuracy
        preds = model.predict(X_valid_fold)
        preds_binary = np.where(preds > 0.5, 1, 0)
        accuracy = accuracy_score(y_valid_fold, preds_binary)
        accuracy_list.append(accuracy)

    return np.mean(accuracy_list)

def objective_catboost(trial):
    param = {
        'verbose': 0,
        'eval_metric': 'Accuracy',
        'loss_function': 'Logloss',
        'iterations': trial.suggest_int('iterations', 100, 1000),  # Increased iterations
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 0.5),
        'depth': trial.suggest_int('depth', 3, 8),
        'l2_leaf_reg': trial.suggest_loguniform('l2_leaf_reg', 1e-8, 1.0),
        'border_count': trial.suggest_int('border_count', 32, 128),
        'bagging_temperature': trial.suggest_loguniform('bagging_temperature', 1e-2, 10.0),
    }

    skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
    accuracy_list = []

    for train_idx, valid_idx in skf.split(X_train, y_train):
        X_train_fold, X_valid_fold = X_train.iloc[train_idx], X_train.iloc[valid_idx]
        y_train_fold, y_valid_fold = y_train.iloc[train_idx], y_train.iloc[valid_idx]

        model = CatBoostClassifier(**param)
        model.fit(X_train_fold, y_train_fold, eval_set=(X_valid_fold, y_valid_fold), verbose=False)

        preds = model.predict(X_valid_fold)
        accuracy = accuracy_score(y_valid_fold, preds)
        accuracy_list.append(accuracy)

    return np.mean(accuracy_list)

# Perform hyperparameter optimization with Optuna

study_xgb = optuna.create_study(direction='maximize')
study_xgb.optimize(objective_xgb, n_trials=100, n_jobs=-1)

study_lgb = optuna.create_study(direction='maximize')
study_lgb.optimize(objective_lgb, n_trials=100, n_jobs=-1)

study_catboost = optuna.create_study(direction='maximize')
study_catboost.optimize(objective_catboost, n_trials=100, n_jobs=-1)

# Print best parameters and best scores for each model
print('XGBoost - Number of finished trials:', len(study_xgb.trials))
print('Best XGBoost trial:', study_xgb.best_trial.params)

print('LightGBM - Number of finished trials:', len(study_lgb.trials))
print('Best LightGBM trial:', study_lgb.best_trial.params)

print('CatBoost - Number of finished trials:', len(study_catboost.trials))
print('Best CatBoost trial:', study_catboost.best_trial.params)

# Train the final models with the best hyperparameters

best_params_xgb = study_xgb.best_trial.params
model_xgb = xgb.XGBClassifier(**best_params_xgb, use_label_encoder=False)
model_xgb.fit(X_train, y_train)

best_params_lgb = study_lgb.best_trial.params
model_lgb = lgb.LGBMClassifier(**best_params_lgb)
model_lgb.fit(X_train, y_train)

best_params_catboost = study_catboost.best_trial.params
model_catboost = CatBoostClassifier(**best_params_catboost)
model_catboost.fit(X_train, y_train, verbose=False)

# Optionally, train GBM model with best hyperparameters if it was included

# Combine models into a Voting Classifier ensemble
voting_classifier = VotingClassifier(estimators=[
    ('xgb', model_xgb),
    ('lgb', model_lgb),
    ('catboost', model_catboost)
], voting='soft')  # Soft voting for probability averaging

# Fit the Voting Classifier on the training data
voting_classifier.fit(X_train, y_train)

# Evaluate on the test set
y_pred_voting = voting_classifier.predict(X_test)
accuracy_voting = accuracy_score(y_test, y_pred_voting)

print(f'Accuracy on test set with Voting Classifier Ensemble: {accuracy_voting:.4f}')
```

```python
import pandas as pd

# Load and preprocess the data
batsman = pd.read_csv("/content/663e2b548c98c_batsman_level_scorecard.csv")
batsman['match_dt'] = pd.to_datetime(batsman['match_dt'])
batsman = batsman.sort_values(by='match_dt')

match_data = pd.read_csv('/content/663e2b6d54457_train_data_with_samplefeatures.csv')
match_data['match_dt'] = pd.to_datetime(match_data['match_dt'])

# Create columns for batsman statistics
num_batsmen = 12  # Adjust according to your maximum number of batsmen
for i in range(1, num_batsmen + 1):
    match_data[f"Batsman_{i}_strike_rate"] = pd.Series([0] * len(match_data))
    match_data[f"Batsman_{i}_weighted_boundaries"] = pd.Series([0] * len(match_data))
    match_data[f"Batsman_{i}_custom_score"] = pd.Series([0] * len(match_data))

# Iterate through each row in match_data
for j in range(match_data.shape[0]):
    roster_ids = match_data["team1_roster_ids"][j]
    match_date = match_data["match_dt"][j]
    batsman_id_list = roster_ids.split(':')
    i = 1
    for batsman_id in batsman_id_list:
        batsman_id = float(batsman_id)

        # Filter the batsman dataset for the current batsman and match date less than the current match date
        filtered_batsmen = batsman[(batsman['batsman_id'] == batsman_id) & (batsman['match_dt'] < match_date)]

        if not filtered_batsmen.empty:
            # Sort filtered DataFrame by 'match_dt' descending
            filtered_batsmen = filtered_batsmen.sort_values(by='match_dt', ascending=False)

            # Calculate the average strike rate over the last 5 matches (if available)
            if filtered_batsmen.shape[0] >= 5:
                last_5_rows = filtered_batsmen.iloc[:5, :]
                avg_strike_rate = last_5_rows['strike_rate'].mean()
            else:
                avg_strike_rate = filtered_batsmen['strike_rate'].mean()

            # Calculate the weighted sum of 4's and 6's over all previous matches
            total_fours = filtered_batsmen['Fours'].sum()
            total_sixes = filtered_batsmen['Sixes'].sum()
            weighted_boundaries = (0.4 * total_fours) + (0.6 * total_sixes)

            # Define the custom score combining the average strike rate and weighted boundaries
            custom_score = avg_strike_rate + weighted_boundaries

            # Store the calculated scores in match_data
            string1 = f"Batsman_{i}_strike_rate"
            match_data.loc[j, string1] = avg_strike_rate
            string2 = f"Batsman_{i}_weighted_boundaries"
            match_data.loc[j, string2] = weighted_boundaries
            string3 = f"Batsman_{i}_custom_score"
            match_data.loc[j, string3] = custom_score
        i += 1
```
